##Cleaning and preprocessing

```python
import numpy as np
import pandas as pd
import sklearn
```

```python
dataset=pd.read_csv('/content/Womens Clothing E-Commerce Reviews.csv')
dataset.head(10)
```

```python
x=dataset.iloc[:, 4].dropna()
Y=dataset.iloc[:, 5]
```

```python
x = x.sample(300)
x.shape
```

```python
y = Y.sample(300)
y.shape
```

##Installing and Applying BERT

```python
X=list(X)
```

```python
import transformers

bert_model = transformers.BertModel.from_pretrained('bert-base-uncased')
```

```python
import numpy as np
import pandas as pd
from transformers import BertModel, BertTokenizer

# Load the BERT model
bert_model = BertModel.from_pretrained('bert-base-uncased')

# Load the BERT tokenizer
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')


# Tokenize the text data
X = bert_tokenizer(X, padding=True, truncation=True, return_tensors='pt')

# Extract features from the BERT model
features = bert_model(X['input_ids']).pooler_output

# Detach the gradient from the features tensor
features = features.detach()

# Train a linear regression model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(features, y)

# Evaluate the linear regression model
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, y, test_size=0.25)

model.score(X_test, y_test)

# Make predictions on new data
new_reviews = ['This product is amazing! I love it!', 'This product is terrible. I do not recommend it.']

new_reviews_tokenized = bert_tokenizer(new_reviews, padding=True, truncation=True, return_tensors='pt')

new_reviews_features = bert_model(new_reviews_tokenized['input_ids']).pooler_output

# Detach the gradient from the new_reviews_features tensor
new_reviews_features = new_reviews_features.detach()

predictions = model.predict(new_reviews_features)

print(predictions)
```
